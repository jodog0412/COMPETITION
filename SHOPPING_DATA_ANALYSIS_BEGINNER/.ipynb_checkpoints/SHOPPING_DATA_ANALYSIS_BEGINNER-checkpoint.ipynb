```python
import pandas as pd
import numpy as np

# 데이터 로드
train = pd.read_csv('dataset/train.csv')
test = pd.read_csv('dataset/test.csv')
sample_submission = pd.read_csv('dataset/sample_submission.csv')

train.info()
```

```python
# 데이터 결측치 처리
train=train.fillna(0)

# Date 전처리
## Date 칼럼에서 "월"에 해당하는 정보만 추출하여 숫자 형태로 반환하는 함수를 작성합니다.
def get_month(date):
    month = date[3:5]
    month = int(month)
    return month
## 이 함수를 Date 칼럼에 적용한 Month 칼럼을 만들어줍니다.
train['Month'] = train['Date'].apply(get_month)

train
```

```python
#test에 대한 전처리

## 결측치 처리
test = test.fillna(0)
## Date 전처리
test['Month'] = test['Date'].apply(get_month)

## isHoliday column에서의 true, false는 1과 0으로 자동처리됨.

train
```

```python
import torch
import torch.nn as nn
from torch import optim

# 분석할 의미가 없는 칼럼을 제거합니다.
train = train.drop(columns=['id','Date'])
test = test.drop(columns=['id','Date'])

# 학습에 사용할 정보와 예측하고자 하는 정보를 분리합니다.
x_train = train.drop(columns=['Weekly_Sales'])
y_train = train[['Weekly_Sales']]

model = nn.Sequential(
    nn.Linear(68805, 34402), # hidden_layer3 = 68805, output_layer = 34402
    nn.ReLU(),
    nn.Linear(34402, 17201), # hidden_layer3 = 34402, output_layer = 17201
    nn.ReLU(),
    nn.Linear(17201, 8600), # hidden_layer3 = 17201, output_layer = 8600
    nn.ReLU(),
    nn.Linear(8600, 6225) # hidden_layer3 = 8600, output_layer = 6225
)
X = torch.tensor(x_train, dtype=torch.float32)
Y = torch.tensor(y_train, dtype=torch.float32)
loss_fn = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters())
losses = []
for epoch in range(100):
  optimizer.zero_grad()
  y_pred = model(X) # forwar 연산
  loss = loss_fn(y_pred, Y)
  loss.backward()
  optimizer.step()

  if epoch % 10 == 0:
    print('Epoch {:4d}/{} Cost: {:.6f}'.format(
            epoch, 100, loss.item()
        ))

  losses.append(loss.item())
```
